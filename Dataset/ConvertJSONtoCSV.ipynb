```python
import json
with open('intents_2.json','r') as f:
    data = json.load(f)
```

```python
from pandas import json_normalize 
d = json_normalize(data, 'intents')
```

```python
d.head()
```

```python
df = d.assign(pattern=d.patterns.str.split(",")).explode('patterns')
```

```python
df.head()
```

```python
df['tags'] = df.tag.apply(lambda x: ', '.join([str(i) for i in x]))
```

```python
df.head()
```

```python
df.drop(['pattern'], axis=1)
```

```python
df.to_csv('intent.csv')
```
